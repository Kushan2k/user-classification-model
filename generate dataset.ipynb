```python
test.loc[:1,:].add('sdfsdf',axis=1,fill_value=10)
```

```python
def assign_values(df:pd.DataFrame,value_list:list,column_name:str)->pd.DataFrame:

    for index, row in df.iterrows():
      df.loc[index, column_name] = random.choice(value_list)
    
    return df
    
```

```python
assign_values(test,Climate_preference,'Climate_preference')
```

```python
assign_values(test,Preferred_cuisines,'Preferred_cuisines')
```
